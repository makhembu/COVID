```python
#!/usr/bin/env python
# coding: utf-8

# In[ ]:


from bs4 import BeautifulSoup
import requests


# In[ ]:


URL = "https://www.worldometers.info/coronavirus/country/kenya/"


# In[ ]:


parsed_html = requests.get(URL)
soup = BeautifulSoup(parsed_html.content, "html.parser")


# In[ ]:


scripts = soup.find_all('script')
script = scripts[21].text
script1 = scripts[22].text
script2 = scripts[23].text
script3 = scripts[24].text
script4 = scripts[25].text


# In[ ]:


import json
dates = script.split("categories: [",1)[1].split("]",1)[0]
dates = "["+dates+"]"
dates = json.loads(dates)
print(dates)


# In[ ]:


total = script.split("data: [",1)[1].split("]",1)[0]
total = "["+total+"]"
total = json.loads(total)
print(total)

daily = script1.split("data: [",1)[1].split("]",1)[0]
daily = "["+daily+"]"
daily = json.loads(daily)
print(daily)

active = script2.split("data: [",1)[1].split("]",1)[0]
active = "["+active+"]"
active = json.loads(active)
print(active)

tdeath = script3.split("data: [",1)[1].split("]",1)[0]
tdeath = "["+tdeath+"]"
tdeath = json.loads(tdeath)
print(tdeath)

ddeath = script4.split("data: [",1)[1].split("]",1)[0]
ddeath = "["+ddeath+"]"
ddeath = json.loads(ddeath)
print(ddeath)

```

```python
from openpyxl import load_workbook
import pandas as pd
writer = pd.ExcelWriter('covidke.xlsx', engine='openpyxl') 
wb  = writer.book
df = pd.DataFrame({'Dates': dates,
                  'Daily New Cases in Kenya': daily,
                   'Total Coronavirus Cases in Kenya': total,
                   'Active Cases in Kenya': active,
                   'Total Coronavirus Deaths in Kenya': tdeath,
                   'Daily New Deaths in Kenya': ddeath})

df.to_excel(writer, index=False)
wb.save('covidke.xlsx')
```
